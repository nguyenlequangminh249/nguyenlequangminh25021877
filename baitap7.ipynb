```python
# tìm kiếm nhị phân 
a = list(map(int,input().split()))
x = int(input())
def timkiem(listx,trai,phai,cantim):
    while trai<=phai:
        i = (trai+phai)//2
        if listx[i]==cantim:
            return i
        if listx[i]<cantim:
            trai+=1
        if listx[i]>cantim:
            phai-=1
    return -1
timkiem(a,0,len(a)-1,x)
```

```python
#ptu xuất hiện nhiều nhất, cùng số lần xuất hiện max thì in cái xh sớm
a = list(map(int,input().split()))
l ={}
def timmax(a):
    for i in a:
        l[i] = l.get(i,0) + 1
    maxs = max(l.values())
    for i in a:
        if l[i]==maxs:
            return i,maxs
print(timmax(a))
```

```python
#cho 1 mảng , hỏi dãy tăng nghiêm ngặt max dài bn
def maxnghiemngat(a):
    b = [1]*len(a)
    for i in range(len(a)):
        for j in range(i):
            if a[j]<a[i]:
                b[i] = max(b[i],b[j]+1)
    return max(b)

maxnghiemngat(list(map(int,input().split())))
```

```python
#khoảng thời gian và truy vấn
def timtruyvan(khoang,truyvan):
    kq=[]
    for i in truyvan:
        mins= float('inf')
        for a,b in khoang:
            if a<=i<=b:
                lens= b-a+1
                if lens<mins:
                    mins=lens
        if mins==float('inf'):
            kq.append(-1)
        else:
            kq.append(mins)
        
    return kq
a= eval(input())
b=eval(input())
timtruyvan(a,b)
        
                
```

```python
#độ dài lớn nhất dãy hài hòa
def longest_harmonious_subsequence(arr):
    from collections import Counter
    
    freq = Counter(arr)
    max_len = 0
    
    for x in freq:
        if x + 1 in freq:
            max_len = max(max_len, freq[x] + freq[x + 1])
    
    return max_len
```

```python
#nhập số các số,nhập các số đó, tìm thứ tự 7 theo chiều từ lớn đến bé
n = int(input())
a = list(map(int, input().split()))

positions = []
for i in range(n):
    if a[i] == 7:
        positions.append(i)

if not positions:
    print("Not found")
else:
    positions.sort(reverse=True)
    print(" ".join(map(str, positions)))
```
