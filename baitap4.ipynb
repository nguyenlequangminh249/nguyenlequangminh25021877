```python
n=int(input())

def lencollat(n):
    step=0
    while n!=1:
        if n%2==0:
            n=n//2
            step+=1
        else:
            n=3*n+1
            step+=1
    return step
maxs = lencollat(1)
for i in range(1,n+1):
    if lencollat(i)>maxs:
        maxs=lencollat(i)
mins=min(i for i in range(1,n+1) if lencollat(i)==maxs)
print(mins,maxs)
```

```python
def thanthiet(a,b):
    c=[i for i in range(1,a) if a%i==0]
    d=[i for i in range(1,b) if b%i==0]
    if sum(c)==b and sum(d)==a:
        return True
    return False

a,b=map(int,input().split())
print(thanthiet(a,b))
```
