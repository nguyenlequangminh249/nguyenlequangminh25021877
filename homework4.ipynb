```python
#bài 3
a = int(input("Nhập số nguyên dương a: "))

print(f"Bảng cửu chương của {a}:")
for i in range(1, 10):
    print(f"{a} x {i} = {a * i}")
```

```python
#bài 4
h = int(input("Nhập chiều cao h: "))

for i in range(1, h + 1):
    # In khoảng trắng phía trước mỗi hàng
    print(" " * (h - i), end="")

    # Hàng đầu và hàng cuối
    if i == 1:
        print("*")
    elif i == h:
        print("*" * (2 * h - 1))
    else:
        print("*" + " " * (2*i -3) + "*")
```

```python
#bài 5
n = int(input("Nhập số nguyên dương n: "))
S = n * (n + 1) // 2
print("Tổng S =", S)
```

```python
#bài 6
a = int(input("Nhập số nguyên dương a: "))

print(f"Các ước dương của {a} là:")
for i in range(1, a + 1):
    if a % i == 0:
        print(i, end=" ")
```
