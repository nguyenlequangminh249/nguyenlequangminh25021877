```python
def merge(a,b):
    c=[]
    i=j=0
    while i<len(a) and j<len(b):
        
        if a[i]<b[j]:
          c.append(a[i])
          i+=1
        else:
          c.append(b[j])
          j+=1
    return c
a=[1,3,5]
b=[2,4,6]
print(merge(a,b))
```

```python
def belong(a,b):
    n=len(a)
    for i in range(len(b)):
        if a==b[i:i+n]:
            return True
    return False
a=[1,2,3,4]
b=[2,3,1,2,3,4,5,6]
print(belong(a,b))

```

```python
def fineradi(a,b):
    c=[]
    d=[]
    i=0
    while i<len(a):
        
        for j in range(len(b)):
            
            c.append(abs(a[i]-b[j]))
        d.append(min(c))
        c=[]
        i+=1
    return max(d)
a=[104231]
b=[43,123,412123,4,2,32,23,90]
print(fineradi(a,b))
    
    
        
        
```
