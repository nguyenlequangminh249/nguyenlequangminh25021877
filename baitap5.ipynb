```python
def maxs(a,b):
    return max(a,b)
print(maxs(1,2))
```

```python
def swap(a,b):
    return (b,a)
print(swap(1,2))
```

```python
def pefe(n):
    if sum(i for i in range(1,n) if n%i==0)==n:
        return True
    return False
print(pefe(6))
```

```python
def changebit(n):
    b=[]
    while n!=0:
        c=n%2
        n=n//2
        b.append(c)
    b.reverse()
    d=''.join(str(i) for i in b)
    return int(d)

def hamm(a,b):
    c=a^b
    bits = str(changebit(c))
    return bits.count('1')
print(hamm(1,4))

        






    
```

```python
def dangcau(a,b):
    c={}
    if len(a)!=len(b):
        return False
    for i in range(len(a)):
        if a[i] not in c:
            c[a[i]]=b[i]
        else:
            if c[a[i]]!=b[i]:
                return False
    return True

print(dangcau('abca','zbxz'))
print(dangcau('abca','zbxd'))
           
            
        
```
