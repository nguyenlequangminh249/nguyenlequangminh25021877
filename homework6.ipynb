```python
def maxl(a):
    maxs=a[0]
    for i in a:
        if i>maxs:
            maxs=i
    return maxs

a = list(map(int,input().split()))
print(a.index(maxl(a)))
        
```

```python
def timvitri(a,k):
    for i in a:
        if i == k:
            print(a.index(i))
            break
        if k not in a:
            print(-1)
            break

a = list(map(int,input().split()))
k= int(input())
timvitri(a,k)
```

```python
def xoabe(x):
    b=[]
    a=list(map(int,input().split()))
    for i in a:
        if i>=x:
            b.append(i)
    return b
            
    
x= int(input())
print(xoabe(x))
```

```python
def chuyen0(a):
    b=[]
    c=[]
    for i in a:
        if i==0:
            b.append(i)
        else:
            c.append(i)
    return b+c
a=list(map(int,input().split()))
print(chuyen0(a))
```
