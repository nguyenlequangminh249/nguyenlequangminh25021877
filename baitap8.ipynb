```python
from collections import Counter

a = list(map(int, input().split()))

cnt = Counter(a)   
xoa = 0           

for x, c in cnt.items():
    if x == 0:
        
        xoa += c
    else:
        if c < x:
            
            xoa += c
        elif c > x:
            
            xoa += c - x
        

if xoa == 0:
    print("YES")
else:
    print("NO", xoa)
```

```python
def merge_sorted_arrays(a, b):
    n, m = len(a), len(b)
    i = j = 0
    result = []

    # Xác định thứ tự sắp xếp
    increasing = True
    if n >= 2 and a[0] > a[1]:
        increasing = False

    while i < n and j < m:
        if increasing:
            if a[i] <= b[j]:
                result.append(a[i])
                i += 1
            else:
                result.append(b[j])
                j += 1
        else:
            if a[i] >= b[j]:
                result.append(a[i])
                i += 1
            else:
                result.append(b[j])
                j += 1

    # Thêm phần còn lại
    while i < n:
        result.append(a[i])
        i += 1

    while j < m:
        result.append(b[j])
        j += 1

    return result

a= list(map(int,input().split()))
b= list(map(int,input().split()))
print(merge_sorted_arrays(a, b))
```

```python
n = int(input())
count = {}

for _ in range(n):
    full_name = input().strip()
    name = full_name.split()[-1]   # lấy tên (từ cuối)
    
    if name in count:
        count[name] += 1
    else:
        count[name] = 1

# số lần xuất hiện lớn nhất
max_freq = max(count.values())

# các tên có số lần xuất hiện = max_freq
candidates = []
for name in count:
    if count[name] == max_freq:
        candidates.append(name)

# lấy tên nhỏ nhất theo thứ tự từ điển
print(min(candidates))
```

```python
s = input().strip()  
k = int(input())

n = len(s)
remove = n - k       
stack = []

for ch in s:
    while stack and remove > 0 and stack[-1] < ch:
        stack.pop()
        remove -= 1
    stack.append(ch)


result = stack[:k]

print("".join(result))
```

```python
s = input()

stack = []

pair = {
    ')': '(',
    '}': '{',
    ']': '[',
    '>': '<'
}

ok = True

for ch in s:
    if ch in "({[<":
        stack.append(ch)
    else:
        if not stack:
            ok = False
            break
        top = stack.pop()
        if pair[ch] != top:
            ok = False
            break

if stack:
    ok = False

if ok:
    print("YES")
else:
    print("NO")
```

```python
def is_composite(x):
    if x <= 3:
        return False
    for i in range(2, int(x**0.5) + 1):
        if x % i == 0:
            return True
    return False


def longest_composite_sequence(N):
    best = []
    current = []

    for i in range(1, N + 1):
        if is_composite(i):
            current.append(i)
        else:
            if len(current) > len(best):
                best = current
            current = []

    # kiểm tra dãy cuối
    if len(current) > len(best):
        best = current

    return best
print(longest_composite_sequence(int(input())))
```

```python
 #Nhập dãy số
a = list(map(int, input().split()))
n = len(a)

# dp[i]: độ dài LIS kết thúc tại i
dp = [1] * n

# prev[i]: lưu vị trí đứng trước của a[i]
prev = [-1] * n

# Tính dp
for i in range(n):
    for j in range(i):
        if a[j] < a[i] and dp[j] + 1 > dp[i]:
            dp[i] = dp[j] + 1
            prev[i] = j

# Tìm vị trí kết thúc của LIS dài nhất
max_len = max(dp)
end = dp.index(max_len)  # lấy dãy xuất hiện sớm nhất

# Truy vết lại dãy
lis = []
while end != -1:
    lis.append(a[end])
    end = prev[end]

lis.reverse()

# In kết quả
print(*lis)
print(max_len)
```

```python
def min_number_by_one_swap(s):
    a = list(s)
    n = len(a)

    for i in range(n):
        min_digit = a[i]
        pos = -1

        # tìm chữ số nhỏ nhất bên phải
        for j in range(i + 1, n):
            if a[j] < min_digit:
                # tránh tạo số có 0 ở đầu
                if i == 0 and a[j] == '0':
                    continue
                min_digit = a[j]
                pos = j
            # nếu bằng nhau, lấy vị trí xa hơn
            elif a[j] == min_digit and pos != -1:
                pos = j

        if pos != -1:
            a[i], a[pos] = a[pos], a[i]
            break  # chỉ đổi đúng 1 lần

    return ''.join(a)
min_number_by_one_swap(input())
```

```python
def can_transform(a, b):
    n, m = len(a), len(b)
    dp = [[False] * (m + 1) for _ in range(n + 1)]
    
    dp[0][0] = True
    
    # a -> rỗng
    for i in range(1, n + 1):
        if a[i-1].islower():
            dp[i][0] = dp[i-1][0]
    
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if a[i-1].isupper():
                if a[i-1] == b[j-1]:
                    dp[i][j] = dp[i-1][j-1]
            else:
                # xóa
                dp[i][j] = dp[i-1][j]
                # đổi thành hoa
                if a[i-1].upper() == b[j-1]:
                    dp[i][j] |= dp[i-1][j-1]
    
    return "YES" if dp[n][m] else "NO"


Q = int(input())
for _ in range(Q):
    a = input().strip()
    b = input().strip()
    print(can_transform(a, b))
```
