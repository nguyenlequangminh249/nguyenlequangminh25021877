```python

numbers = list(map(int, input().split()))
k = int(input())


pairs = []


for i in range(len(numbers)):
    for j in range(len(numbers)):  
        if  numbers[i] + numbers[j] == k:
            pairs.append((numbers[i], numbers[j]))


pairs.sort(key=lambda x: x[0])

print(pairs)
```

```python

numbers = list(map(int, input().split()))


frequency = {}


for num in numbers:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1


most_frequent = min(frequency, k=lambda x: (-frequency[x], x))


print(most_frequent)
```
